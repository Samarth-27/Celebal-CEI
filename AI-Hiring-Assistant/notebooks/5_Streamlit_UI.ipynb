# Streamlit Demo

**Goal:** Create a user-friendly web interface that ties together all the modules (Parsing, Classification, Semantic Matching, XAI, and Conversational AI) into an end-to-end Hiring Assistant application.

## Why this module exists
A machine learning pipeline is only as good as its usability. We need a frontend to demonstrate how candidates and recruiters will interact with the system. Streamlit provides a fast, Python-native way to build this UI.

## How it integrates
- **Recruiter View**: Uploads JDs, views ranked candidates, and sees the semantic matching scores.
- **Candidate View**: Uploads a resume, gets parsed data, sees their Explainable AI feedback report, and can chat with the AI Mentor using the RAG pipeline.

## Implementation
We will write the Streamlit application to `../app.py` from this notebook. To run it, use:
`streamlit run app.py` from the root directory.
